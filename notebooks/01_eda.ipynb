# Exploratory Data Analysis - Fake Job Post Detection

This notebook performs exploratory data analysis on the fake job postings dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
```

```python
# Load the dataset
df = pd.read_csv('../data/fake_job_postings.csv')
print(f"Dataset shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nData types:\n{df.dtypes}")
```

```python
# Basic statistics
print("Basic Statistics:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Target variable distribution
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='fraudulent')
plt.title('Distribution of Real vs Fake Jobs')
plt.xlabel('Fraudulent (0 = Real, 1 = Fake)')
plt.ylabel('Count')

# Add percentages
total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.show()
```

```python
# Correlation heatmap for numeric features
numeric_cols = ['telecommuting', 'has_company_logo', 'has_questions', 'fraudulent']
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
# Top industries
plt.figure(figsize=(12, 6))
industry_counts = df['industry'].value_counts().head(10)
sns.barplot(x=industry_counts.values, y=industry_counts.index)
plt.title('Top 10 Industries')
plt.xlabel('Count')
plt.show()
```

```python
# Fraudulent jobs by industry
fraud_by_industry = df[df['fraudulent'] == 1]['industry'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=fraud_by_industry.values, y=fraud_by_industry.index)
plt.title('Top 10 Industries with Fake Jobs')
plt.xlabel('Number of Fake Jobs')
plt.show()
```

```python
# Text length analysis
df['description_length'] = df['description'].fillna('').apply(len)
df['title_length'] = df['title'].fillna('').apply(len)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Description length
sns.histplot(data=df, x='description_length', hue='fraudulent', 
             ax=ax1, alpha=0.7, bins=50)
ax1.set_title('Description Length Distribution')
ax1.set_xlabel('Description Length')
ax1.set_xlim(0, 5000)

# Title length
sns.histplot(data=df, x='title_length', hue='fraudulent', 
             ax=ax2, alpha=0.7, bins=30)
ax2.set_title('Title Length Distribution')
ax2.set_xlabel('Title Length')

plt.tight_layout()
plt.show()
```

```python
# Employment type distribution
plt.figure(figsize=(10, 6))
employment_fraud = pd.crosstab(df['employment_type'], df['fraudulent'], normalize='index') * 100
employment_fraud.plot(kind='bar', stacked=True)
plt.title('Employment Type vs Fraudulent Jobs')
plt.xlabel('Employment Type')
plt.ylabel('Percentage')
plt.legend(['Real', 'Fake'], title='Job Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Location analysis (top 20 locations)
location_counts = df['location'].value_counts().head(20)

plt.figure(figsize=(12, 8))
sns.barplot(x=location_counts.values, y=location_counts.index)
plt.title('Top 20 Job Locations')
plt.xlabel('Number of Jobs')
plt.show()
```
